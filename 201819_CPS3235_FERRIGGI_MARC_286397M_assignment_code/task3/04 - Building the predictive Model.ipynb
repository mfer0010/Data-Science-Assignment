```python
%reset

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
```

# ANOCOVA Regression

```python
#read data
data = pd.read_pickle('./data/properties.pkl')
data.info()
```

```python
#drop nulls
data = data.dropna()
data.info()
```

```python
#First 200 readings are the training set, the other 37 will be the testing set
training = data[:200]
testing = data[200:]
```

## Convert all categorical variables to dummy variables and remove one of the columns 

```python
garage = pd.get_dummies(training['garage'],drop_first=True, dtype=float)
location = pd.get_dummies(training['location'],drop_first=True, dtype=float)
prop_type = pd.get_dummies(training['type'],drop_first=True, dtype=float)
pool = pd.get_dummies(training['pool'],drop_first=True, dtype=float)
```

```python
X = np.c_[training['area'],garage,location,prop_type,pool]
y = training['price']
```

```python
X.shape[1]
```

## Estimate the Regression Parameters

```python
model = sm.OLS(y,X).fit()
model.params
```

```python
model.summary()
```

The majority of the predictors are not significant, thus remove location

```python
X = np.c_[training['area'],garage,prop_type,pool]
y = training['price']
model = sm.OLS(y,X).fit()
model.summary()
```

```python
X
```

## Fit the testing set

```python
params = model.params
```

```python
garage = pd.get_dummies(testing['garage'],drop_first=True, dtype=float)
location = pd.get_dummies(testing['location'],drop_first=True, dtype=float)
prop_type = pd.get_dummies(testing['type'],drop_first=True, dtype=float)
pool = pd.get_dummies(testing['pool'],drop_first=True, dtype=float)
```

```python
X = np.c_[testing['area'],garage,prop_type,pool]
true_y = testing['price']
```

```python
true_y = true_y.reset_index(drop=True)
true_y[1]
```

```python
error = abs(true_y-np.dot(X,params))
error
```

```python
np.mean(error)
```

## Checking the Error with full model

```python
garage = pd.get_dummies(training['garage'],drop_first=True, dtype=float)
location = pd.get_dummies(training['location'],drop_first=True, dtype=float)
prop_type = pd.get_dummies(training['type'],drop_first=True, dtype=float)
pool = pd.get_dummies(training['pool'],drop_first=True, dtype=float)

X = np.c_[training['area'],garage,location,prop_type,pool]
y = training['price']
model = sm.OLS(y,X).fit()

params = model.params

garage = pd.get_dummies(testing['garage'],drop_first=True, dtype=float)
location = pd.get_dummies(testing['location'],drop_first=True, dtype=float)
prop_type = pd.get_dummies(testing['type'],drop_first=True, dtype=float)
pool = pd.get_dummies(testing['pool'],drop_first=True, dtype=float)
X = np.c_[testing['area'],garage,location,prop_type,pool]
true_y = testing['price']
true_y = true_y.reset_index(drop=True)

error = abs(true_y-np.dot(X,params))
np.mean(error)
```
