```python
%reset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Required by Jupyter for inlining images
%matplotlib inline
```

```python
#Load the Employment Rate Data
employment_rate = pd.read_csv('./data/tesem010.tsv',sep='\t', header=None).transpose()
#Set the First Row as headers
headers = employment_rate.iloc[0]
employment_rate  = pd.DataFrame(employment_rate.values[1:], columns=headers)

employment_rate.head()
```

```python
#Find Malta's Data
print(headers[1:40])
```

```python
malta_columns = [0,29,68,107]
maltese_data = employment_rate.iloc[:,malta_columns]
maltese_data.columns = ['Year','Females','Males','Total']
maltese_data.head()
```

```python
maltese_data.info()
```

```python
#Remove b's from data
maltese_data.set_value(0,'Females',34.8) 
maltese_data.set_value(0,'Males',79.7)
maltese_data.set_value(0,'Total',57.4)
```

```python
#change values to numeric
maltese_data = maltese_data.apply(pd.to_numeric)
maltese_data.info()
```

```python
fig = plt.figure()
plt.plot(maltese_data['Year'],maltese_data['Females'])
plt.plot(maltese_data['Year'],maltese_data['Males'])
plt.plot(maltese_data['Year'],maltese_data['Total'], '--')
plt.xlabel('Year')
plt.ylabel('Employment Rate /%')
plt.title('Employment Rate Time Series for Malta')
plt.legend()
plt.grid(1,which='major',axis='y')
fig.savefig('./plots/er_time_series.png',dpi=300)
```

```python
#Get 2017 Data of females for rank plot
#First get totals:
data_2017 = employment_rate.iloc[:,1:40].iloc[12,:]
data_2017
```

```python
data_2017[1] = 63.6
data_2017[7] = 73.7
```

```python
data_2017 = pd.to_numeric(data_2017)
```

```python
#Index for plots
index = [['Austria','Belgium', 'Bulgaria','Croatia','Cyprus','Czechia','Germany','Denmark','Euro Area (18 Countries)',
         'Euro Area (19 Countries)','Estonia','Greece','Spain','EU (27 Countries)','EU (28 Countries)','Finland','France',
         'France, Metropolitan','Croatia','Hungary','Ireland','Iceland','Italy','Lithuania','Luxembourg','Latvia','Montenegro',
         'Macedonia','Malta','Netherlands','Norway','Poland','Portugal','Romania','Sweden','Slovenia','Slovakia','Turkey','UK']]
index.append(data_2017.tolist())
data_2017 = pd.DataFrame(index).transpose()
data_2017
#data_2017 = pd.DataFrame(data_2017)
#data_2017 = data_2017.transpose()
```

```python
#sort list
data_2017 = data_2017.sort_values(by=[1,0],ascending=False)
data_2017 = data_2017.reset_index(drop=True)
data_2017
```

```python
ax = data_2017[28:].plot.barh(x=0,y=1,color='#73A9E8')

# Despine
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)

# Switch off ticks
ax.tick_params(axis='both', which='both', bottom=0, top=0, labelbottom=1, left=0, right=0, labelleft=1)

#Remove the legend
ax.get_legend().remove()

ax.set_xlabel('Employment Rate /%')
ax.set_ylabel('Country')
ax.set_title('10 Worst Countries for Female Employment Rate in the EU',size=16)

ax.get_figure().savefig('./plots/er_bar_chart.png',bbox_inches='tight')
```
